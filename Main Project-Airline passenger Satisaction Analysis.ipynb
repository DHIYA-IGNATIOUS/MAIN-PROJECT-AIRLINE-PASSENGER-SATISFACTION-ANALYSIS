# 2.Data Pre-processing

```python
1.Missing values if any
```

```python
print("\nüîπ Missing Values in Each Column:")
print(df.isnull().sum())
```

2.Clearing Missing values

```python
train_df['Arrival Delay in Minutes'] = train_df['Arrival Delay in Minutes'].fillna(0)
print("‚úÖ Missing values cleared")
```

3.Ignore warnings

```python
import warnings
warnings.filterwarnings('ignore')
```

4.Removing Duplicates 

```python
duplicate_rows = train_df[train_df.duplicated()]
```

```python
print("Number of duplicate rows:", duplicate_rows.shape[0])
```

5.Formating and standardisation 

```python
cat_cols = ['Gender','Customer Type','Type of Travel','Class','satisfaction']

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

df['Gender'] = df['Gender'].replace({'Male':'M','Female':'F'})

print("‚úÖ Categorical columns formatted!")
```

6.Checking Unique values

```python
for col in cat_cols:
    print(col, ":", df[col].unique())
```

7.Derived Features

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0,25,35,50,100],
    labels=['Young','Adult','Middle-Age','Senior']
)

df['Delay_Status'] = np.where(
    df['Arrival Delay in Minutes'] > 15, 'Delayed', 'On Time'
)

print("Derived features created successfully!")
```

8.Convert to Categorical Data Types

```python
cat_final = ['Gender','Customer Type','Type of Travel','Class','satisfaction','Age_Group','Delay_Status']

for col in cat_final:
    df[col] = df[col].astype('category')

print(df[cat_final].dtypes)
```

# 3. Exploratory Data Analysis (EDA)

# UNIVARIATE ANALYSIS(Distribution & counts)

(a) Age Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'].dropna(), bins=20, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

print("Insight: Most passengers are in age ranges shown on the histogram (often 25-45).")
```

(b)Satisfaction ratio (pie) and counts

```python
print("\nSatisfaction counts:")
print(df['satisfaction'].value_counts())
plt.figure(figsize=(5,5))
df['satisfaction'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Passenger Satisfaction Ratio')
plt.ylabel('')
plt.show()
print("Insight: See percent of Satisfied vs Neutral/Dissatisfied ‚Äî indicates overall customer sentiment.")
```

# BIVARIATE ANALYSIS(groupby, orderby, aggregation)

(a)Satisfaction by Customer Type (counts)

```python
cust_satis = df.groupby(['Customer Type','satisfaction']).size().unstack(fill_value=0)
display(cust_satis)
```

(b)Sorted by total satisfied count

```python
cust_total = cust_satis.copy()
cust_total['Total_satisfied'] = cust_total.get('Satisfied', 0)  # adjust label if your dataset differs
cust_total = cust_total.sort_values(by='Total_satisfied', ascending=False)
display(cust_total)

print("Insight: Loyal customers tend to have higher satisfied counts than disloyal customers.")
```

(c)Average flight distance by Class (groupby + sort)

```python
avg_distance = df.groupby('Class')['Flight Distance'].mean().sort_values(ascending=False)
display(avg_distance)

print("Insight: Business class typically has higher average flight distance (long-haul passengers).")
```

PIVOT TABLE

Pivot: Flight Distance mean by Gender and Class

```python
pivot_distance = pd.pivot_table(df, values='Flight Distance', index='Gender', columns='Class', aggfunc='mean')
display(pivot_distance)
print("Insight: Pivot table compares average distances across gender and class ‚Äî helpful to spot differences.")
```

# MULTIVARIATE (catplot / cross-tab)

```python
sns.catplot(x='Customer Type', hue='satisfaction', col='Class', data=df, kind='count', height=4, aspect=0.9)
plt.suptitle('Customer Type vs Class vs Satisfaction', y=1.02)
plt.show()
print("Insight: Loyal + Business class subset shows highest satisfaction and economy + disloyal shows lower satisfaction.")
```
